```python
import pandas as pd
from sklearn.model_selection import train_test_split

train = pd.read_csv('datasets/house_prices/train.csv')
test = pd.read_csv('datasets/house_prices/test.csv')
target = 'SalePrice'
X = train.drop(columns=target)
y = train[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from xgb_tuning import tune_xgb_regressor
xgb_results, _ = tune_xgb_regressor(
    X_train,
    y_train,
    X_test,
    y_test,
    target_col='SalePrice',
    n_trials = 20,
    n_splits = 5,
    # param_space: Optional[Dict[str, Dict[str, Any]]] = None,
)
```

```python
from xrfm_tuning import tune_xrfm_regressor
xrfm_results, _ = tune_xrfm_regressor(
    X_train,
    y_train,
    X_test,
    y_test,
    target_col='SalePrice',
    n_trials = 20,
    n_splits = 5,
    # param_space: Optional[Dict[str, Dict[str, Any]]] = None,
)
```

```python
from mlp_tuning import tune_mlp_regressor

mlp_results, preprocessor = tune_mlp_regressor(
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    target_col="SalePrice",
    n_trials=20,
    n_splits=5,
)

print(mlp_results)
```
